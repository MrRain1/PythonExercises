```python
import pandas as pd
import matplotlib.pyplot as plt

data_file = pd.read_csv("./2.1 QueryResults.csv", header= 0, names = ["Date", "Tag", "Posts"])
data_file.head()

# convert date to datatime:
data_file["Date"] = pd.to_datetime(data_file["Date"])
```

```python
data_file.tail()
```

```python
data_file.count()
```

```python
data_file.groupby("Tag").sum()
```

```python
data_file.groupby("Tag").count()
```

```python
reshaped_df = data_file.pivot(index= "Date", columns="Tag", values= "Posts")
reshaped_df
```

```python
reshaped_df.shape
```

```python
reshaped_df.head()
```

```python
reshaped_df.tail()
```

```python
# Dealing with NaN values using .fillna() method:
reshaped_df.fillna(0, inplace=True)
reshaped_df
```

```python
reshaped_df.isna().values.any()
```

```python
plt.figure(figsize =(16,10))
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.xlabel("Date", fontsize = 14)
plt.ylabel("Number of posts", fontsize = 14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index, reshaped_df.java)
plt.plot(reshaped_df.index, reshaped_df["python"])
```

```python
# plotting multiple data on the same chart:
plt.figure(figsize =(16,10))
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.xlabel("Date", fontsize = 14)
plt.ylabel("Number of posts", fontsize = 14)
plt.ylim(0,35000)
for column in reshaped_df.columns:
    plt.plot(reshaped_df.index, reshaped_df[column], linewidth = 3, label= reshaped_df[column].name)
    
plt.legend(fontsize = 16)
```

```python
roll_df = reshaped_df.rolling(window=6).mean()

plt.figure(figsize =(16,10))
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.xlabel("Date", fontsize = 14)
plt.ylabel("Number of posts", fontsize = 14)
plt.ylim(0,35000)
for column in roll_df.columns:
    plt.plot(roll_df.index, roll_df[column], linewidth = 3, label= roll_df[column].name)
    
plt.legend(fontsize = 16)
```
